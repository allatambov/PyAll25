### Задача 1

Питон Петя написал на заборе `python + anaconda = Love ♡`, а мы сохранили эту надпись в переменную `on_the_fence`.

Исследуйте работу методов `.lower()`, `.upper()`, `.capitalize()`, `.title()` на примере этой строки.

```python
on_the_fence = "python + anaconda = Love ♡"

print(on_the_fence.lower()) # все маленькими буквами
print(on_the_fence.upper()) # все заглавными буквами
print(on_the_fence.capitalize()) # первая буква в строке – заглавная
print(on_the_fence.title()) # все первые буквы у слов – заглавные
```

### Задача 2

Напишите программу, которая запрашивает имя и фамилию пользователя через пробел, и:

* выводит на экран сообщение *Принято!*, если слова, разделенные пробелами, написаны с заглавной буквы;
* иначе выводит на экран сообщение *Проверьте написание!*

**Пример работы программы:**

Ввод:
    
    Скарапея Питонова
    
Вывод:

    Принято!
    
Ввод:

    аня Самсонова
    
Вывод:
    
    Проверьте написание!

```python
# вариант 1

data = input()
if data == data.title():
    print("Принято!")
else:
    print("Проверьте написание!")
```

```python
# вариант 2

data = input()
if data.istitle():
    print("Принято!")
else:
    print("Проверьте написание!")
```

### Задача 3

В переменной `text` сохранены фрагменты песни из фильма «Ученик лекаря» (музыка М.Таривердиева, слова В.Коростылева). Знаки пунктуации при этом убраны. Давайте считать, что питон Петя из задачи №1 вышел на новый уровень и теперь поет Анаконде песни.

```python
text = """
Два озера твои глаза
Я в них окончу путь
Я просто выбрать не могу
В каком мне утонуть
Я просто выбрать не могу
И я покуда жив
Но я как в песне соловей
Перед тобой не лжив
Вот перед ней вот перед ней
Вот перед ней какую трель рассыпал менестрель
Твои глаза твои глаза
Высоких два окна
И люди говорят что в них
Любовь ко мне видна
И я на цыпочки привстал
Чтоб выше стать чуть-чуть
Я только выбрать не могу
В какое заглянуть
"""
```

a. Напишите программу, которая сохранит в переменную `words` список всех слов *без повторений* с сортировкой по алфавиту. Все слова при этом должны быть написаны с маленькой буквы.

Фрагмент ожидаемого списка:

    ['в', 'видна', 'вот', 'выбрать', 'высоких', 'выше', 'глаза', 'говорят', ...]

b. Напишите программу, которая посчитает, сколько раз каждое слово из `words` встречается в тексте `text` и сохранит в переменную `counts` пары *слово - соответствующая ему частота*.

Фрагмент ожидаемого списка:

    [('в', 20), ('видна', 1), ('вот', 1), ('выбрать', 3), ('высоких', 0), ('выше', 1), ...]

*Сначала разберем пункт а)*

```python
# вариант 1: цикл с условием + функция sorted()

parts = text.lower().split()
words = []
for p in parts:
    if p not in words:
        words.append(p)
words = sorted(words)
print(words)
```

```python
# вариант 2: цикл с условием + метод .sort()

parts = text.lower().split()
words = []
for p in parts:
    if p not in words:
        words.append(p)
words.sort()
print(words)
```

```python
# вариант 3: без цикла, но с множеством
# функция set() создает множество – перечень элементов без дубликатов,
# а функция sorted() после сортировски всегда возвращает список

parts = text.lower().split()
words = sorted(set(parts))
print(words)
```

*Теперь пункт b)*

```python
counts = []
for word in words:
    freq = text.count(word) # сколько раз слово word встречается в text
    counts.append((word, freq))
print(counts)
```
